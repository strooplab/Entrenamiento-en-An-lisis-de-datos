# Muertes por coronavirus
Análisis de datos de un dataset con reportes de muertes por coronavirus en el 2021

# Librerias

```python
import pandas as pd
import os
```

# Directorio

```python
path = os.getcwd()
directorio = os.path.join(path, 'datasets/')

if not os.path.exists(directorio):
    os.makedirs(directorio)
```

## Datasets

```python
path, dirs, files = next(os.walk(directorio))
file_count = len(files)
if file_count != 0:
    df_list = []

    for i in range(file_count):
        if files[i].endswith('.csv'):
            temp_df = pd.read_csv(directorio + files[i], 
                                  encoding='utf-8'
                                  )
        elif files[i].endswith('.xlsx'):
            temp_df = pd.read_excel(directorio + files[i], 
                                    engine='openpyxl',
                                    )
        else:
            print(f"File format or encoding from {files[i]} is not supported")
        df_list.append(temp_df)
else:
    print("No files found in the directory")
```
